The Rectified Linear Unit (ReLU) is a piecewise linear activation function defined as f(x)=max(0,x), which outputs the input directly if it is positive and zero otherwise. It is the default choice for hidden layers in deep neural networks due to its computational efficiency and ability to mitigate the vanishing gradient problem, allowing models to train faster and perform better than with older functions like sigmoid or tanh. 

```python
def ReLU(x):
    return max(0,x)
```

```python
ReLU(-10)
```

```python
ReLU(10)
```

```python
def Leaky_ReLU(x):
    return max(0.1*x,x)
```

```python
Leaky_ReLU(100)
```

```python
Leaky_ReLU(-100)
```

```python
Leaky_ReLU(-7)
```
