```python
import numpy as np
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
def mse(y_true, y_predicted):
    total_error=0
    for yt, yp in zip(y_true,y_predicted):
        total_error+=(yt-yp)**2
    print("totalr error",total_error)
    mse=(total_error)/len(y_predicted)
    print("MSE",mse)
    return total_error
    
```
