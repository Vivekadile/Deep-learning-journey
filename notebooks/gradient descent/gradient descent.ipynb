```python
def sigmoid(x):
    import math
    return 1/(1+math.exp(-x))
    
```

```python
def los_loss(y_true,y_predicted):
    epsilon=1e-15
    y_predicted_new=[max(i,epsilon) for i in y_predicted]
    y_predicted_new=[min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new=np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(age,afforadbility,y_true ,epochs ,loss_thresold):
    w1=w2=1
    bias=0
    rate=0.5
    n=len(age)

    for i in range(epochs):
        weighted_sum=w1*age+w2*afforadbility+bias
        y_predicted=sigmoid_numpy(weighted_sum)
        
        loss=los_loss(y_true,y_predicted)

        w1d= np.dot(np.transpose(age),(y_predicted-y_true))
        w2d= np.dot(np.transpose(afforadbility),(y_predicted-y_true))

        bias_d=np.mean(y_predicted-y_true)

        w1 = w1-rate*w1d
        w2 = w2-rate*w2d

        print(f"Epoch:{i},w1:{w1},w2{w2},bias:{bias},loss:{loss}")
        if loss<=loss_thresold:
            break
            

    return w1,w2,bias
        
```
