```python
import matplotlib.pyplot as plt
import numpy as np

# Data
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

# Training Data
lstm_train = [0.9851, 0.9855, 0.9950, 0.9895]
cnn_train = [0.9751, 0.9765, 0.9876, 0.9814]
lr_train = [0.7950, 0.7770, 0.9432, 0.8519]

# Test Data
lstm_test = [0.5763, 0.6243, 0.7169, 0.6669]
cnn_test = [0.5536, 0.6102, 0.6834, 0.6444]
lr_test = [0.7466, 0.7571, 0.8962, 0.8124]

x = np.arange(len(metrics))
width = 0.25

# Plot
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Training Graph
ax[0].bar(x - width, lstm_train, width, label='LSTM')
ax[0].bar(x, cnn_train, width, label='CNN')
ax[0].bar(x + width, lr_train, width, label='LR')

ax[0].set_title('Training Metrics')
ax[0].set_xticks(x)
ax[0].set_xticklabels(metrics)
ax[0].legend()

# Test Graph
ax[1].bar(x - width, lstm_test, width, label='LSTM')
ax[1].bar(x, cnn_test, width, label='CNN')
ax[1].bar(x + width, lr_test, width, label='LR')

ax[1].set_title('Test Metrics')
ax[1].set_xticks(x)
ax[1].set_xticklabels(metrics)
ax[1].legend()

plt.tight_layout()
plt.show()
```
